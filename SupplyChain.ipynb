```python
def deseasonalized_forecast(alpha, initial_forecast, demand, deseason_demand):
    deseason_forecast = [initial_forecast]
    for i in range(len(demand)-1):
        deseason_forecast.append(alpha*deseason_demand[i]+(1-alpha)*deseason_forecast[i])
    return deseason_forecast
```

```python
###initial_forcast should be a list
def moving_average_forecast(weight_of_average, initial_forecast, demand):
    b = weight_of_average
    deseason_forecast = []
    deseason_forecast = initial_forecast
    for i in range(len(demand)-b):
        x = np.sum(demand[i:(i+b)])
        z = x/b
        deseason_forecast.append(z)
       
    return deseason_forecast
```

```python
def forecastHolt(period, alpha, beta, resultA):
    n = len(resultA)
    P = 0
    Q = 0
    for i in range(n):
        P = P + (i)*resultA[i]
        Q = Q + resultA[i]
    Sxy = n*P - ((n*(n-1))/2)*Q
    Sxx = (n*n*(n+1)*(2*n+1))/6 - (n*n*(n+1)*(n+1))/4
    b = Sxy/Sxx
    R = np.mean(resultA)
    a = R - (b*(n+1))/2
    S = [a]
    G = [b]
    for i in range(0, n):
        A = alpha*resultA[i] + (1 - alpha)*(S[i] + G[i])
        S.append(A)
        B = beta*(S[i+1] - S[i]) + (1 - beta)*G[i]
        G.append(B)
    tau = period
    FHolt = []
    for i in range(len(S)):
        FHolt.append(S[i] + tau*G[i])
    
    return FHolt[:-1]
```
